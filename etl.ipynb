```python
import pandas as pd
from etl import preprocess_data  # Importa la función desde etl.py
```

```python
# Carga los archivos CSV de reservas y hoteles
df_book = pd.read_csv('data/bookings_train.csv')
df_hotel = pd.read_csv('data/hotels.csv')

# Une los dos datasets por la columna 'hotel_id'
df = pd.merge(df_book, df_hotel, on='hotel_id')

# Llama a la función preprocess_data
df_preprocessed, df_validation_preprocessed = preprocess_data(df)
```

```python
print("DataFrame preprocesado:")
df_preprocessed
```

```python
# Matriz de correlación
corr_matrix = df_preprocessed.corr()

# Filtrar solo las correlaciones de la variable 'cancelled_last_30_days'
corr_cancelled_last_30_days = corr_matrix[['cancelled_last_30_days']].sort_values(by='cancelled_last_30_days', ascending=False)

# Convertir las correlaciones a un DataFrame
correlation_df = corr_cancelled_last_30_days.reset_index()

# Imprimir el DataFrame de correlaciones
print(correlation_df)
```

```python
# Matriz de correlación
corr_matrix = df_preprocessed.corr()

# Filtrar solo las correlaciones de la variable 'cancelled_last_30_days'
corr_cancelled_last_30_days = corr_matrix['cancelled_last_30_days']

# Filtrar correlaciones positivas y negativas
positive_correlations = corr_cancelled_last_30_days[corr_cancelled_last_30_days > 0]
negative_correlations = corr_cancelled_last_30_days[corr_cancelled_last_30_days < 0]

# Sumar las correlaciones positivas y restar 1
positive_sum = positive_correlations.sum() - 1

# Sumar las correlaciones negativas en valor absoluto
negative_sum = negative_correlations.abs().sum()

# Normalizar las sumas
total_sum = positive_sum + negative_sum
normalized_positive_sum = positive_sum / total_sum
normalized_negative_sum = negative_sum / total_sum

# Imprimir los resultados
print(f"Suma de correlaciones positivas menos 1: {positive_sum}")
print(f"Suma de correlaciones negativas en valor absoluto: {negative_sum}")
print(f"Suma normalizada de correlaciones positivas: {normalized_positive_sum}")
print(f"Suma normalizada de correlaciones negativas: {normalized_negative_sum}")
```
